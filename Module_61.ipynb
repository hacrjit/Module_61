### <b>Question No. 12</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Google Play Store dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
df = pd.read_csv(url)

# Replace 'Varies with device' with NaN
df['Size'] = df['Size'].replace('Varies with device', np.nan)

# Convert the 'Size' column to numeric
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', '')) * 1000000
    elif 'k' in size:
        return float(size.replace('k', '')) * 1000
    else:
        return np.nan

df['Size'] = df['Size'].apply(lambda x: convert_size(x) if isinstance(x, str) else x)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Size'], df['Rating'], alpha=0.5)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')
plt.show()
```

### <b>Question No. 13</b>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
df = pd.read_csv(url)

# Remove dollar signs from the 'Price' column and convert to numeric
df['Price'] = df['Price'].apply(lambda x: float(x.replace('$', '')) if '$' in x else 0)

# Group by 'Category' and calculate the average price for each category
avg_price_by_category = df.groupby('Category')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(14, 6))
avg_price_by_category.plot(kind='bar', color='skyblue')
plt.xlabel('App Category')
plt.ylabel('Average Price ($)')
plt.title('Average Price by App Category')
plt.xticks(rotation=90)
plt.show()
```

### <b>Question No. 14</b>

```python
import pandas as pd

# Load the Google Play Store dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
df = pd.read_csv(url)

# Remove any non-numeric characters from the 'Installs' column and convert to integer
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').fillna(0).astype(int)

# Create a frequency table of the 'App' column
app_frequency = df['App'].value_counts().reset_index()
app_frequency.columns = ['App', 'Frequency']

# Sort the frequency table by 'Frequency' in descending order
app_frequency_sorted = app_frequency.sort_values(by='Frequency', ascending=False)

# Display the top 10 most popular apps
top_10_apps = app_frequency_sorted.head(10)
print(top_10_apps)
```

### <b>Question No. 15</b>

To identify the most popular app categories on the Google Play Store and make recommendations to the company, we can follow these steps:

1. **Load and Explore the Dataset**: Load the Google Play Store dataset and explore its contents to understand the available features and data types.

2. **Clean the Data**: Clean the dataset by handling missing values, removing duplicates, and converting relevant columns to the appropriate data types.

3. **Analyze App Categories**: Analyze the 'Category' column to determine the distribution of apps across different categories.

4. **Calculate Popularity**: Calculate the popularity of each app category based on the number of installs or ratings.

5. **Visualize the Data**: Create visualizations such as bar charts or pie charts to present the distribution of apps across categories and the popularity of each category.

6. **Make Recommendations**: Based on the analysis, identify the most popular app categories and make recommendations to the company regarding the potential success of launching a new app in those categories.

7. **Consider Other Factors**: Consider other factors such as competition, market trends, and user demographics when making recommendations.

Here's a basic example of how we could approach this task:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
df = pd.read_csv(url)

# Clean the 'Installs' column by removing commas and plus signs, and converting to numeric
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').fillna(0)

# Calculate the total number of installs per category
installs_by_category = df.groupby('Category')['Installs'].sum()

# Create a bar chart to visualize the popularity of each category
plt.figure(figsize=(12, 6))
installs_by_category.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.xlabel('App Category')
plt.ylabel('Total Installs')
plt.title('Total Installs by App Category')
plt.xticks(rotation=90)
plt.show()

# Identify the most popular app categories based on the total number of installs
most_popular_categories = installs_by_category.sort_values(ascending=False).head(5)
print("Most Popular App Categories:")
print(most_popular_categories)
```

In this example, we load the dataset, clean the 'Installs' column, calculate the total number of installs per category, and visualize the popularity of each category using a bar chart. We then identify the most popular app categories based on the total number of installs.
